```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_excel('/content/india trade data.xlsx')
data
```

```python
# # data.rename(columns={'Unnamed: 0':'slno'},inplace=True)
# data["Trade Worth (USD)"]=data["Trade Worth (USD)"].str.replace(' '," ")
# # data["Trade Worth (USD)"]=data["Trade Worth (USD)"].astype(int
# # data["Trade Worth (USD)"]=data["Trade Worth (USD)"].astype(int)
# data
data[['Trade Value', 'Currency']] = data['Trade Worth (USD)'].str.split(' ', expand=True)

# Remove commas and dollar signs from the 'Trade Value' column
data['Trade Value'] = data['Trade Value'].str.replace(',', '').str.replace('$', '')

# Convert 'Trade Value' to numeric
data['Trade Value'] = pd.to_numeric(data['Trade Value'])

# Display the updated DataFrame
data
```

```python
data.drop(columns='Currency',axis=1,inplace=True)
data.drop(columns='Trade Worth (USD)',axis=1,inplace=True)
data
```
